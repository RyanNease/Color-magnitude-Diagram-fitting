```python
import numpy as np
import matplotlib.pyplot as plt

#Part A


R =5.0 # radius of the half-pipe in meters
g =9.8 #graviational constant m/s^2
dt = .01 #timestep
num_steps = 1000 #number of steps taken
av=0 #inital angular velocity

phi = 1 # 1 radian

time = np.zeros(num_steps) #create a time array of zeros
position = np.zeros(num_steps) #create a position array of zeros

#create a for loop to update the time array and position array
for i in range(num_steps):
    time[i]= i*dt # update the time array with each time step
    aac = (-g/R)*np.sin(phi) # angular accelaration equation
    phi = phi + av*dt + (aac*dt**2)/2 # calculate the new position
    aac1 = (-g/R)*np.sin(phi) #given differential equation
    av = av + (dt/2)*(aac +aac1)# solve for new velocity
    position[i]= phi # update the array with the new position

plt.plot(time,position) #create a plot of time and position
plt.xlabel('Time') #label x axis time
plt.ylabel('Position') #label y axis position
plt.title('Angle vs Time') #title is angle vs time
```

```python
#Part B



R =5.0 # radius of the half-pipe in meters
g =9.8 #graviational constant m/s^2
dt = .01 #timestep
num_steps = 1000 #number of steps taken
av=0 #inital angular velocity

phi =(20*np.pi)/180   # phi in radians
phiinitial = (20*np.pi)/180 # initial value of phi that wont change

time = np.zeros(num_steps) #create a time array of zeros
position = np.zeros(num_steps) #create a time array of position

for i in range(num_steps):
    time[i]= i*dt # update the time array by each time step


    #exact solution
    aac = (-g/R)*np.sin(phi) # angular accelaration equation
    phi = phi + av*dt + (aac*dt**2)/2 # calculate the new position
    aac1 = (-g/R)*np.sin(phi) #given differential equation
    av = av + (dt/2)*(aac +aac1)# solve for new velocity
    position[i]= phi # update the array with the new position


#finding the approximate solution
def approxsolution(x):
    phiapprox = phiinitial*np.cos(np.sqrt(g/R)*x)
    return phiapprox


plt.plot(time,position,label='exact solutions') #plot exact solutions
plt.plot(time, approxsolution(time), label ='approx solutions') #plot aprroximate solutions
plt.xlabel('Time') #label x axis time
plt.ylabel('Position') # label y axis position
plt.title('Angle vs Time') # title the graph angle vs time
plt.legend() #list the legend
```

Graph b
As we can see when we use an inital angle that is relatively small such as 20 degrees our approximate solution and exact solution line up rather nicely. For the small angles the reason the two solutions are so similar is because the skateboards motion on the halfpipe will closely resemble a simple harmonic motion. From time 0 to around 5 they are practicly identical and then at that point we start to see some slight deviation between the two. Thats because as time goes on the difference between the linear approximation and the actual behavior of sin(phi) becomes more noticeable. While linear approximations represent the fundamental simple harmonic motion, real oscilating systems deal with higher harmonics that come into play as the time passes. In addition the actual motion of the skateboard is nonlinear due to the prescense of the sin(phi). Also the phase of the skateboards motion can shift over time. These three things could be the reason for why as time passes the exact and approximate solutions begin to differ.

Graph c
Once we start to use angles that are bigger, such as pi/2, the approximate solutions and the exact solutions are vastly different. From the start of this graph the two solutions begin to vary and as the graph goes on that difference continues to widen. From these graphs we can conclude that the approximate solutions will only work for small angles, and more specifically for a short period of time. At larger angles the motion of the system no longer follows the simple harmonic motion as the change in direction will become sharper causing a larger difference between the exact solution and the approximate one that increases over time.
